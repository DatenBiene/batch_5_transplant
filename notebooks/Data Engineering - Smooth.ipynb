## Smoothing Dynamic Set

```python
import pandas as pd
from transplant.config import * 
from transplant.visualization.graph import * 
```

Dans ce notebook nous réalisons une simple fonction smoothing sur le set dynamique. Nous mesurons également le temps d'exécution du script. Pour un rolling mean sur 5 périodes, le script s'éxecute en ~12 secondes. 

```python
df = pd.read_csv(PATH_DYNAMIC_CLEAN)
df_static = pd.read_csv(PATH_STATIC_CLEAN)
```

```python
df = df[DYNAMIC_HEADERS]
```

```python
import time
start = time.time()
melt = pd.melt(df, id_vars = ['id_patient', 'time']).groupby(['id_patient', 'variable']).rolling(5, min_periods =1, on='time').value.mean().reset_index()
end = time.time()
print(end - start)
```

```python
import time
start = time.time()
melt.pivot_table(index=['id_patient', 'time'], columns='variable', values='value').reset_index()
end = time.time()
print(end - start)
```

```python
dates = df_static[['id_patient', 'date_transplantation', 'date_sortie_bloc']]
```

```python
a = pd.merge(df, dates, on='id_patient')
```

```python
a = a[['id_patient', 'time','date_sortie_bloc']]
```

```python
a['date_sortie_bloc'] = pd.to_datetime(a['date_sortie_bloc'])
a['time'] = pd.to_datetime(a['time'])
```

```python
a['diff'] = a['date_sortie_bloc'] - a['time']
```

```python
a.groupby('id_patient')['diff'].min().sort_values(ascending=True).reset_index()
```
