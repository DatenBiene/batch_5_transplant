```python
import pandas as pd
import numpy as np
import sys
import pandas_profiling


```

```python
fct_haemo=pd.read_csv('data/model/fct_haemodynamic.csv')
dim_patient=pd.read_csv('data/dim_patient.csv')
#df_merge=pd.merge(df,dim_patient[['id_patient','date_transplantation']], on='id_patient')
#df_merge['timestamp']=df_merge['date_transplantation']+_+df_merge['time']
df = pd.merge(fct_haemo, dim_patient[['id_patient','date_transplantation']], on='id_patient')
df['timestamp'] = df['date_transplantation'] + '-' +df['time']
df = df[df['time'] != '0']
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
print(dim_patient)
```

```python
profile=pandas_profiling.ProfileReport(df)
```

```python
profile.to_file(outputfile="myoutputfile.html")
```

```python
ids=list(set(list(df['id_patient'])))
print(ids)

for i in ids:
    table=df[df['id_patient']==i]
    profile=pandas_profiling.ProfileReport(table)
    profile.to_file(outputfile="./patients_haemo/myoutputfile"+str(i)+".html")
```
